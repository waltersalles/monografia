```python
import pandas as pd

# Dados de infla√ß√£o de 1985 a 1994
inflacao_pre_real = {
    "Ano": [1985, 1986, 1987, 1988, 1989, 1990, 1991, 1992, 1993, 1994],
    "Infla√ß√£o (%)": [225.99, 147.14, 228.34, 629.11, 1430.72, 2947.73, 432.79, 951.96, 1927.38, 2075.89]
}

df_pre_real = pd.DataFrame(inflacao_pre_real)

# Exibir tabela
print(df_pre_real)
```

üìä Tabela 1: Taxas de infla√ß√£o anuais (1985‚Äì1994)
üìç Inserir logo ap√≥s o subt√≥pico 2.7 ‚ÄúRaz√µes do fracasso‚Äù

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df_pre_real["Ano"], df_pre_real["Infla√ß√£o (%)"], marker='o', color='darkred')
plt.title("Infla√ß√£o Anual no Brasil (1985‚Äì1994)")
plt.xlabel("Ano")
plt.ylabel("Infla√ß√£o (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

üìà Gr√°fico: Infla√ß√£o anual (1985‚Äì1994)

```python
# Dados de infla√ß√£o e c√¢mbio no in√≠cio do Plano Real
real_dados = {
    "Ano": [1994, 1995, 1996, 1997, 1998, 1999, 2000],
    "Infla√ß√£o (%)": [2075.89, 66.01, 15.76, 6.93, 3.20, 4.86, 7.04],
    "C√¢mbio (BRL/USD)": [None, 0.92, 1.01, 1.08, 1.16, 1.82, 1.83]
}

df_real = pd.DataFrame(real_dados)

# Exibir tabela
print(df_real)
```

üìä Tabela 2: Taxas de infla√ß√£o e c√¢mbio (1994‚Äì2000)

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.set_title("Infla√ß√£o e C√¢mbio no Brasil (1994‚Äì2000)")
ax1.set_xlabel("Ano")
ax1.set_ylabel("Infla√ß√£o (%)", color='tab:red')
ax1.plot(df_real["Ano"], df_real["Infla√ß√£o (%)"], color='tab:red', marker='o', label='Infla√ß√£o')
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx()
ax2.set_ylabel("C√¢mbio (BRL/USD)", color='tab:blue')
ax2.plot(df_real["Ano"][1:], df_real["C√¢mbio (BRL/USD)"][1:], color='tab:blue', marker='s', label='C√¢mbio')
ax2.tick_params(axis='y', labelcolor='tab:blue')

fig.tight_layout()
plt.grid(True)
plt.show()
```

üìà Gr√°fico: Infla√ß√£o e c√¢mbio (1994‚Äì2000)

```python
# Dados do per√≠odo Dilma
dilma_dados = {
    "Ano": [2010, 2011, 2012, 2013, 2014, 2015, 2016],
    "Infla√ß√£o (%)": [5.04, 6.64, 5.40, 6.20, 6.33, 9.03, 8.74],
    "C√¢mbio (BRL/USD)": [1.76, 1.67, 1.95, 2.16, 2.35, 3.34, 3.48]
}

df_dilma = pd.DataFrame(dilma_dados)

# Exibir tabela
print(df_dilma)
```

üìä Tabela 3: Infla√ß√£o e c√¢mbio (2010‚Äì2016)

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.set_title("Infla√ß√£o e C√¢mbio no Brasil (2010‚Äì2016)")
ax1.set_xlabel("Ano")
ax1.set_ylabel("Infla√ß√£o (%)", color='tab:red')
ax1.plot(df_dilma["Ano"], df_dilma["Infla√ß√£o (%)"], color='tab:red', marker='o', label='Infla√ß√£o')
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx()
ax2.set_ylabel("C√¢mbio (BRL/USD)", color='tab:blue')
ax2.plot(df_dilma["Ano"], df_dilma["C√¢mbio (BRL/USD)"], color='tab:blue', marker='s', label='C√¢mbio')
ax2.tick_params(axis='y', labelcolor='tab:blue')

fig.tight_layout()
plt.grid(True)
plt.show()
```
